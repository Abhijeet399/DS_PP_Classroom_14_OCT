```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

```python
X = np.arange(10)
```

```python
X
```

```python
y1 = 20 * X + 100
y2 = X**2
```

```python
plt.plot(X, y1, color = 'red', label = 'St Line', linestyle = '--')
plt.plot(X, y2, color = 'blue', label = 'Parabola')
plt.xlabel('height')
plt.ylabel('weight')
plt.legend(title = 'Types of curve:', bbox_to_anchor=(0.9, 0.9999))
```

```python
plt.scatter(X, y1, c = 'red', label = 'St Line', marker = '*')
plt.scatter(X, y2, c = 'blue', label = 'Parabola', marker = '^')
plt.xlabel('height')
plt.ylabel('weight')
plt.legend(title = 'Types of curve:', bbox_to_anchor=(0.9, 0.9999))
```

```python
plt.bar(['English', 'Maths', 'Phy', 'EVS'], [90, 85, 95, 100])
plt.show()
```

```python
pos = np.array([1, 4, 7])
plt.bar(pos, [90, 85, 100])
# pos = np.array([2, 5, 8])
plt.bar(pos + 0.80, [80, 90, 95])
plt.show()
```

```python
pos = np.array([1, 4, 7])
plt.bar(pos, [90, 85, 100], width = 0.5)
# pos = np.array([2, 5, 8])
plt.bar(pos + 0.50, [80, 90, 95], width=0.5, color = 'red')
plt.show()
```

```python
plt.pie([80, 40, 20, 90, 100], labels = ('Eng', 'Phy', 'Hindi', 'Maths', 'EVS'), explode = (0.1, 0.1, 0, 0.11, 0.05), shadow=True, autopct = '%.2f', radius=2)
```
