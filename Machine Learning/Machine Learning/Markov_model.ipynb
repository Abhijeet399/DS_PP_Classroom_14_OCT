```python
import numpy as np
import pandas as pd
import os
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
path = 'C:/Users/bhatt/OneDrive/Desktop/whisper-main/sherlock/'

def read_all_text(path):
    text = []
    
    for _, _, files in os.walk(path):
        for file in files:
            with open(path + file) as f:
                for line in f:
                    line = line.strip()
                    if line == '----------':
                        break
                    if line != '':
                        text.append(line)
    return text
```

```python
all_data = read_all_text(path)
```

```python
len(all_data)
```

```python
def clean_data(all_data):
    clean_text = []
    for line in all_data:
        line = line.lower()
        line = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", '', line)
        words = word_tokenize(line)
        tokens = []
        for w in words:
            tokens.append(w)
        clean_text += words
    return clean_text
```
