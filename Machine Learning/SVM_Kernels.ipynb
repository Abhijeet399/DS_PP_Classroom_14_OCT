```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
data, labels = make_circles(n_samples = 1000, noise = 0.1, factor = 0.2)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = labels)
```

```python
print(data.shape)
```

```python
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
axis_3 = x**2 + y**2
```

```python
dataset = np.hstack((data, axis_3))
```

```python
dataset.shape
```

```python
ax = plt.subplot(projection = '3d')
ax.scatter(x, y, axis_3, c = labels)
plt.show()
```

```python
from sklearn import svm
```

```python
svm_classifier = svm.SVC(kernel = 'linear')
```

```python
svm_classifier.fit(dataset, labels)
```

```python
svm_classifier.score(dataset, labels)
```

```python
weights = svm_classifier.coef_
bias = svm_classifier.intercept_
```

```python
weights, bias
```

```python
mesh_x, mesh_y = np.meshgrid(dataset[:, 0], dataset[:, 1])
```

```python
mesh_x.shape
```

```python
plane = -(weights[0, 0] * mesh_x + weights[0, 1] * mesh_y + bias[0])/weights[0, 2]
```

```python
ax = plt.subplot(projection = '3d')
ax.scatter(x, y, axis_3, c = labels)
ax.plot_surface(mesh_x, mesh_y, plane)
plt.show()
```

```python
ax = plt.subplot(projection = '3d')
ax.scatter(x, y, axis_3, c = labels)
ax.plot_surface(dataset[:, 0], dataset[:, 1], plane)
plt.show()
```
