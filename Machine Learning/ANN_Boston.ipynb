```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import keras
import tensorflow as tf
```

```python
from keras.datasets import boston_housing
```

```python
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
print(train_X.shape, test_X.shape, train_Y.shape, test_Y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
train_X = min_max_scaler.fit_transform(train_X)
test_X = min_max_scaler.fit_transform(test_X)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential(name = 'Boston_ANN')
model.add(Dense(500, input_dim = train_X.shape[1], activation = 'relu', name = 'Input_layer'))
model.add(Dense(400, activation = 'relu', name = 'Layer_2'))
model.add(Dense(300, activation = 'relu', name = 'Layer_3'))
model.add(Dense(200, activation = 'relu', name = 'Layer_4'))
model.add(Dense(100, activation = 'relu', name = 'Layer_5'))
model.add(Dense(1, activation = 'relu', name = 'Output_Layer'))
```

```python
model.compile(optimizer = 'Adam', loss = 'mse', metrics = [keras.metrics.RootMeanSquaredError()])
```

```python
model.summary()
```

```python
hist = model.fit(train_X, train_Y, epochs = 1000, batch_size = 200, validation_data=(test_X, test_Y))
```
