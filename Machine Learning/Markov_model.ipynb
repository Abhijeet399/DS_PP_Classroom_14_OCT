```python
import numpy as np
import pandas as pd
import os
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
path = 'C:/Users/bhatt/OneDrive/Desktop/whisper-main/sherlock/'

def read_all_text(path):
    text = []
    
    for _, _, files in os.walk(path):
        for file in files:
            with open(path + file) as f:
                for line in f:
                    line = line.strip()
                    if line == '----------':
                        break
                    if line != '':
                        text.append(line)
    return text
```

```python
all_data = read_all_text(path)
```

```python
def clean_data(all_data):
    clean_text = []
    for line in all_data:
        line = line.lower()
        line = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", '', line)
        words = word_tokenize(line)
        tokens = []
        for w in words:
            tokens.append(w)
        clean_text += words
    return clean_text
```

```python
n_gram = 2

markov_model = {}

for i in range(len(clean_words) - n_gram- 1):
    curr_word, next_word = "", ""
    for j in range(n_gram):
        curr_word += clean_words[i + j] + " "
        next_word += clean_words[i + j + n_gram] + " "
    curr_word = curr_word[: -1]
    next_word = next_word[: -1]
    
    if curr_word not in markov_model:
        markov_model[curr_word] = {}
        markov_model[curr_word][next_word] = 1
    else:
        if next_word in markov_model[curr_word]:
            markov_model[curr_word][next_word] += 1
        else:
            markov_model[curr_word][next_word] = 1
```

```python
import random

def gen_story(markov_model, start, limit = 7):
    count = 0
    curr_word = start
    next_word = None
    line = curr_word + " "
    
    while (count < limit):
        next_word = random.choices(list(markov_model[curr_word].keys()),
                                  markov_model[curr_word].values())
        count += 1
        line += next_word[0] + " "
        curr_word = next_word[0]
    return line
```

```python
for i in range(10):
    print(gen_story(markov_model, 'my god'))
```

```python
for i in range(10):
    print(gen_story(markov_model, 'i am'))
```
